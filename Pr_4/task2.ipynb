
### Sentiment Analysis with Ukrainian Models


```python

# List of sentiment analysis models
sentiment_models = [
    "cointegrated/rubert-tiny-sentiment-balanced",
    "distilbert/distilbert-base-multilingual-cased"
]
```

```python

# Sample Ukrainian texts
texts = [
    "Це найкращий день у моєму житті!",
    "Мені дуже прикро через ці новини.",
    "Я неймовірно радий бачити старих друзів."
]
```

```python

# Function to perform sentiment analysis
def analyze_sentiment(model_name, texts):
    print(f"Model: {model_name}")
    sentiment_pipeline = pipeline("sentiment-analysis", model=model_name)
    results = sentiment_pipeline(texts)
    for text, result in zip(texts, results):
        print(f"Text: {text}")
        print(f"Sentiment: {result['label']}, Score: {result['score']:.4f}\n")
    print("-" * 50)
```

```python

# Run sentiment analysis for each model
for model in sentiment_models:
    analyze_sentiment(model, texts)
```


### Named Entity Recognition (NER) with Ukrainian Models


```python

# List of NER models
ner_models = [
    "ukr-models/uk-ner",
    "julian-schelb/roberta-ner-multilingual"
]
```

```python

# Sample Ukrainian text for NER
ner_text = "Президент України Володимир Зеленський зустрівся з представниками Європейського Союзу у Києві."
```

```python

# Function to perform NER
def perform_ner(model_name, text):
    print(f"Model: {model_name}")
    ner_pipeline = pipeline("ner", model=model_name, aggregation_strategy="simple")
    entities = ner_pipeline(text)
    for entity in entities:
        print(f"Entity: {entity['word']}, Type: {entity['entity_group']}, Score: {entity['score']:.4f}")
    print("-" * 50)

# Run NER for each model
for model in ner_models:
    perform_ner(model, ner_text)
```


### Text Generation in Ukrainian


```python
# List of text generation models
text_gen_models = [
    "benjamin/gpt2-large-wechsel-ukrainian"
]

# Prompt for text generation
prompt_text = "Україна сьогодні святкує"
```

```python

# Function to generate text using tokenizer and model
def generate_text_with_tokenizer(model_name, prompt, max_length=50):
    from transformers import AutoTokenizer, AutoModelForCausalLM

    print(f"Model: {model_name}")

    # Load tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForCausalLM.from_pretrained(model_name)

    # Tokenize the input prompt
    inputs = tokenizer(prompt, return_tensors="pt")

    # Generate text using the model
    outputs = model.generate(
        inputs['input_ids'],
        max_length=max_length,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        top_k=50,
        top_p=0.95,
        temperature=0.8
    )

    # Decode and print the generated text
    for idx, output in enumerate(outputs):
        generated_text = tokenizer.decode(output, skip_special_tokens=True)
        print(f"Generated Text {idx+1}: {generated_text}\n")
    print("-" * 50)
```

```python

# Run text generation for each model
for model in text_gen_models:
    generate_text_with_tokenizer(model, prompt_text)
```
